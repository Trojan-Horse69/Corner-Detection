```python
import cv2
import numpy as np
```

```python
img = cv2.imread('C:/Users/j/squares2.jpg')
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(img_gray, 500, 0.1, 100) 
#100 = max corners, 0.01 = corner quality to be detected, 10 = minimum distance between each corner
```

```python
corners = np.int0(corners)
```

```python
for corner in corners:
    x, y = corner.ravel()
    cv2.circle(img, (x, y), 5, (0, 0, 255), -1)
    #3 = edges pixel, -1 = thickness
```

REAL TIME CORNER DETECTION

```python
cap = cv2.VideoCapture(0)

#to create trackbar
def nothing(x): #done for memory purposes. If there's no value changing on the trackbar it's poing to pass
    pass


cv2.namedWindow("Frame")
cv2.createTrackbar("quality", "Frame", 1, 100, nothing)

while True:
    _, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    quality = cv2.getTrackbarPos("quality", "Frame") #To get the value of the trackbar in real time
    quality = quality/100 if quality > 0 else 0.01
    corners = cv2.goodFeaturesToTrack(gray, 100, quality, 20)
    
    if corners is not None:
        corners = np.int0(corners)
        
        for corner in corners:
            x, y = corner.ravel()
            cv2.circle(frame, (x, y), 3, (0, 0, 255), -1)
    
    cv2.imshow("Frame", frame)
    
    key = cv2.waitKey(1)
    if key==27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
